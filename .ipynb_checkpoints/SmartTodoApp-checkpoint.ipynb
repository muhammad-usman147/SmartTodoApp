# Preparing the Data

### Converting the data into pair of sentences
### 1 row  ===> 1 sentence

```python
dataToConvert = data.iloc[:,1:-1] #eliminating the date and prac column
```

```python
dataToConvert
```

```python
columns = data.columns[1:] #getting all columns
columns
```

##### the code below is to convert the values into the string

```python
sentence = []
for i in range(len(dataToConvert)):
    x = dataToConvert.iloc[i,:].values
    s_s = ''
    s_s +=" "+str(i+1) #addig the date as from day 0 to day 730
    for index,value in enumerate(x):
        if value == 1.0:
            s_s+=" "+columns[index]
    sentence.append(s_s)
```

```python
sentence[-5:-1]
```

###### Now the data has been converted into sentences, now for the text processing part

```python
import keras
```

```python
sentence[3]
```

```python
MAXS = 0
for x in sentence:
    if len(x) > MAXS:
        MAXS = len(x)
MINS = MAXS
for x in sentence:
    if len(x) < MINS:
        MINS = len(x)
print("The max and min length of the sentence is ==> {} and {}".format(MAXS,MINS))
```

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
```

```python
maxlen = MAXS
maxwords = 1000
```

```python
token = Tokenizer(num_words=maxwords)
```

```python
token.fit_on_texts(sentence)
```

```python
sequences = token.texts_to_sequences(sentence)
```

#### METHOD ====> 2

```python
ON_OFF_sentence = []
for i in range(len(dataToConvert)):
    x = dataToConvert.iloc[i,:].values
    s_s = ''
    s_s +=" "+str(i+1) #addig the date as from day 0 to day 730
    for index,value in enumerate(x):
        if value == 1.0:
            s_s+=" "+'1'
        elif value == 0.0:
            s_s+=" "+'0'
    ON_OFF_sentence.append(s_s)
```

```python
ON_OFF_sentence
```

##### GETTING X and Y

```python
MAXS = 0
for x in ON_OFF_sentence:
    if len(x) > MAXS:
        MAXS = len(x)
MINS = MAXS
for x in ON_OFF_sentence:
    if len(x) < MINS:
        MINS = len(x)
print("The max and min length of the sentence is ==> {} and {}".format(MAXS,MINS))
```

##### Splitting every single character into an array

```python
single_char_arr = []
for _sent in ON_OFF_sentence:
    single_char_arr.append(_sent.split())
```

```python
MAXS = 0
for x in single_char_arr:
    if len(x) > MAXS:
        MAXS = len(x)
MINS = MAXS
for x in single_char_arr:
    if len(x) < MINS:
        MINS = len(x)
print("The max and min length of the sentence is ==> {} and {}".format(MAXS,MINS))
```
