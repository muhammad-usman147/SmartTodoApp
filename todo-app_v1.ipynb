# Preparing the Data

### Converting the data into pair of sentences
### 1 row  ===> 1 sentence

```python
dataToConvert = data.iloc[:,1:-1] #eliminating the date and prac column
```

```python
dataToConvert
```

```python
columns = data.columns[1:] #getting all columns
columns
```

##### the code below is to convert the values into the string

```python
sentence = []
for i in range(len(dataToConvert)):
    x = dataToConvert.iloc[i,:].values
    s_s = ''
    s_s +=" "+str(i+1) #addig the date as from day 0 to day 730
    for index,value in enumerate(x):
        if value == 1.0:
            s_s+=" "+columns[index]
    sentence.append(s_s)
```
