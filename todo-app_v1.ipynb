# Preparing the Data

### Converting the data into pair of sentences
### 1 row  ===> 1 sentence

```python
dataToConvert = data.iloc[:,1:-1] #eliminating the date and prac column
```

```python
dataToConvert
```

```python
columns = data.columns[1:] #getting all columns
columns
```

##### the code below is to convert the values into the string

```python
sentence = []
for i in range(len(dataToConvert)):
    x = dataToConvert.iloc[i,:].values
    s_s = ''
    s_s +=" "+str(i+1) #addig the date as from day 0 to day 730
    for index,value in enumerate(x):
        if value == 1.0:
            s_s+=" "+columns[index]
    sentence.append(s_s)
```

###### Now the data has been converted into sentences, now for the text processing part

```python
import keras
```

# Creating Token

```python
from keras.preprocessing.text import Tokenizer
```

```python
tokens = Tokenizer(num_words=1000)
```

```python
tokens.fit_on_texts(sentence)
```

```python
numerical_data = tokens.texts_to_sequences(sentence)
```

```python
numerical_data
```

# Creating X and Y
### Getting the last numerical text  for the Y value
### Removing the last numerical text from the X arrays

```python
ydata = []
xdata = []
for i in range(len(numerical_data)):
    ydata.append(numerical_data[i][-1])
    # getting the last value from the text
    xdata.append(numerical_data[i][:-1])
    # getting all the values from text while neglecting the last value
```

```python
ydata[0:5]
```

```python
#padding into Sequence
from keras.preprocessing.sequence import pad_sequences
maxlen = []
for i in xdata:
    maxlen.append(len(i))
max_len = max(maxlen)
print("MAXLEN of a sentence is ==> {}".format(max_len))
```

```python
padded_text = pad_sequences(xdata, maxlen=max_len)
```

```python
padded_text[0:3]
```

# Creating the Network

```python
from keras.models import Sequential
from keras.layers import Dense,LSTM,Dropout,Embedding
from keras.optimizers import Adam,rmsprop
from keras import Input,Model
from keras.utils import to_categorical
```

```python
numpy_ydata = to_categorical(ydata)
```

```python
xdata = np.reshape(padded_text,(730,23,1))
```

```python
numpy_ydata
```

```python
padded_text.shape,xdata.shape
```

```python
def BaseLine():
    inpt = Input(shape=(xdata.shape[1],xdata.shape[2]))
    x = LSTM(300)(inpt)
    
    x = Dense(64,activation='relu')(x)
    outpt  = Dense(28,activation='sigmoid')(x)
    model = Model(inpt,outpt)
    model.compile(optimizer=Adam(),loss='categorical_crossentropy',metrics=['accuracy'])
    return model
```

```python
xxmodel = BaseLine()
xxmodel.summary()
```

```python
xxmodel.fit(xdata,numpy_ydata,epochs=10,batch_size=1)
```
